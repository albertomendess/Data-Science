# Relatório de Análise VI
### Criando novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/aluguel_residencial.csv', sep= ';')
dados.head(7)
```

```python
dados = dados.loc[:, ~dados.columns.str.contains('^Unnamed')]
dados
```

```python
dados['Valor_Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head(7)
```

```python
dados['Valor_m²'] = (dados['Valor'] / dados['Area']).round(2)
dados.head(7)
```

```python
dados['Valor_Bruto_M²'] = (dados['Valor_Bruto'] / dados['Area']).round(2)
dados.head(7)
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo_Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados
```

## Excluindo Variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipo_Agregado', 'Valor_m²', 'Valor_Bruto', 'Valor_Bruto_M²']])
dados_aux
```

```python
del dados_aux['Valor_Bruto']
dados_aux
```

```python
dados_aux.pop('Valor_Bruto_M²')
dados_aux
```

```python
dados_aux.drop(['Valor_m²'], axis= 1, inplace= True)
dados_aux
```

```python
dados.drop(['Valor_Bruto', 'Valor_Bruto_M²'], axis= 1, inplace= True)
dados
```

```python
dados.to_csv('/content/aluguel_residencial.csv', sep= ';')
```
